```python
cls2rgb = {cl:list(classes.loc[cl, :]) for cl in classes.index}
idx2rgb={idx:np.array(rgb) for idx, (cl, rgb) in enumerate(cls2rgb.items())}
```

```python
def adjust_mask(mask):  
    semantic_map = []
    for colour in list(cls2rgb.values()):        
        equality = np.equal(mask, colour)
        class_map = np.all(equality, axis = -1)
        semantic_map.append(class_map)
    semantic_map = np.stack(semantic_map, axis=-1)
    return np.float32(semantic_map)
```

```python
def map_class_to_rgb(p):
    return idx2rgb[p[0]]
```

```python
batch_sz=16
seed = 2
BASE_PATH = 'CamVid/'
```

```python
image_datagen = ImageDataGenerator(rescale=1./255)
mask_datagen  = ImageDataGenerator(dtype=int) 
image_generator = image_datagen.flow_from_directory(
    BASE_PATH,
    class_mode=None,
    classes=['train'],
    seed=seed,
    batch_size=batch_sz,
    target_size=(224,224))

mask_generator = mask_datagen.flow_from_directory(
    BASE_PATH,
    classes=['train_labels'],
    class_mode=None,
    seed=seed,
    batch_size=batch_sz,
    color_mode='rgb',
    target_size=(224,224))

train_generator = zip(image_generator, mask_generator)

def train_generator_fn():
    for (img,mask) in train_generator:
        new_mask = adjust_mask(mask)
        yield (img,new_mask)     

val_image_generator = image_datagen.flow_from_directory(
    BASE_PATH,
    class_mode=None,
    classes=['val'],
    seed=seed,
    batch_size=batch_sz,
    target_size=(224,224))

val_mask_generator = mask_datagen.flow_from_directory(
    BASE_PATH,
    classes=['val_labels'],
    class_mode=None,
    seed=seed,
    batch_size=batch_sz,
    color_mode='rgb',
    target_size=(224,224))
val_generator = zip(val_image_generator, val_mask_generator)

def val_generator_fn():
    for (img,mask) in train_generator:
        new_mask = adjust_mask(mask)
        yield (img,new_mask) 
```

```python
test_image_generator = image_datagen.flow_from_directory(
    'CamVid/',
    class_mode=None,
    classes=['test'],
    seed=seed,
    batch_size=batch_sz,
    target_size=(224,224))

test_mask_generator = mask_datagen.flow_from_directory(
    'CamVid/',
    classes=['test_labels'],
    class_mode=None,
    seed=seed,
    batch_size=batch_sz,
    color_mode='rgb',
    target_size=(224,224))

test_generator = zip(test_image_generator, test_mask_generator)

def test_generator_fn():
    for (img,mask) in test_generator:
        new_mask = adjust_mask(mask)
        yield (img,new_mask)  
```

```python
def conv2d_block(input_tensor, n_filters, kernel_size = 3, batchnorm = True):
    x = Conv2D(filters = n_filters, kernel_size = (kernel_size, kernel_size), kernel_initializer = 'he_normal', padding = 'same')(input_tensor)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation('relu')(x)
    return x
```

```python
def unet(n_filters = 16, batchnorm = True, dropout = 0.1, num_classes=32):

    inputs = Input(shape=(224, 224, 3), name="input_image")
    encoder = keras.applications.MobileNetV2(input_tensor=inputs, input_shape=(224, 224, 3), include_top=False)

    skip_connection_names = ["input_image", "block_1_expand_relu", "block_3_expand_relu", "block_6_expand_relu"]
    encoder_output = encoder.get_layer("block_13_expand_relu").output

    x = encoder_output
    x_skip_1 = encoder.get_layer(skip_connection_names[-1]).output 
    x_skip_2 = encoder.get_layer(skip_connection_names[-2]).output 
    x_skip_3 = encoder.get_layer(skip_connection_names[-3]).output 
    x_skip_4 = encoder.get_layer(skip_connection_names[-4]).output 

    u6 = Conv2DTranspose(n_filters * 13 , (3, 3), strides = (2, 2), padding = 'same')(x)
    u6 = concatenate([u6, x_skip_1])
    c6 = conv2d_block(u6, n_filters * 13, kernel_size = 3, batchnorm = batchnorm)
    p6 = Dropout(dropout)(c6)

    u7 = Conv2DTranspose(n_filters * 12, (3, 3), strides = (2, 2), padding = 'same')(c6)
    u7 = concatenate([u7, x_skip_2])
    c7 = conv2d_block(u7, n_filters * 12, kernel_size = 3, batchnorm = batchnorm)
    p7 = Dropout(dropout)(c7)

    u8 = Conv2DTranspose(n_filters * 11, (3, 3), strides = (2, 2), padding = 'same')(p7)
    u8 = concatenate([u8, x_skip_3])
    c8 = conv2d_block(u8, n_filters * 11, kernel_size = 3, batchnorm = batchnorm)
    p7 = Dropout(dropout)(c8)

    u9 = Conv2DTranspose(n_filters * 10, (3, 3), strides = (2, 2), padding = 'same')(c8)
    u9 = concatenate([u9, x_skip_4])
    c9 = conv2d_block(u9, n_filters * 10, kernel_size = 3, batchnorm = batchnorm)
    
    outputs = Conv2D(num_classes, (1, 1), activation='softmax')(c9)

    model = keras.models.Model(inputs=inputs, outputs=outputs)
    
    return model
```

```python
model = unet(n_filters = 16, batchnorm = True, dropout = 0.25, num_classes=num_classes)

encoder_layers = model.layers[0:-22]

for layer in encoder_layers:
    layer.trainable = False

model.summary()
```

```python
smooth=1e-07
def dice_coef(y_true, y_pred):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersect = K.sum(y_true_f * y_pred_f, axis=-1)
    denom = K.sum(y_true_f + y_pred_f, axis=-1)
    return K.mean((2. * intersect / (denom + smooth)))

def dice_loss(y_true, y_pred):
    return 1 - dice_coef(y_true, y_pred)

def cce_dice_loss(y_true, y_pred):
    cce = keras.losses.CategoricalCrossentropy()
    return dice_loss(y_true, y_pred) + cce(y_true, y_pred)

def iou(y_true, y_pred):
    intersection = K.sum(y_true * y_pred)
    sum_ = K.sum(y_true + y_pred)
    jac = (intersection + smooth) / (sum_ - intersection + smooth)
    return jac
```

```python
model_name = 'model-{epoch:03d}-{val_accuracy:.2f}.h5'
save_dir = 'saved_model/'
    
if not os.path.isdir(save_dir):
    os.makedirs(save_dir)

filepath = os.path.join(save_dir, model_name)

checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath, verbose=1, monitor='val_loss',save_best_only=True, mode='auto')  
lr_reducer = tf.keras.callbacks.ReduceLROnPlateau(factor=0.6, cooldown=0, patience=2,
                             monitor='val_loss',verbose=1)
early_stopping = keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=3, mode='max')


callbacks = [lr_reducer, checkpoint, early_stopping]

optimizer=tf.keras.optimizers.Adam(lr=1e-4)

metrics = [dice_coef, iou, 'accuracy']
```

```python
model.compile(loss=cce_dice_loss, optimizer=optimizer, metrics=metrics)
```

```python
history = model.fit_generator(train_generator_fn(),
                    validation_data=val_generator_fn(),
                    steps_per_epoch=len(train)//batch_sz,
                    validation_steps=len(val)//batch_sz,
                    epochs=20,
                    callbacks=callbacks)
```

```python
model.evaluate(test_generator_fn(), steps=len(test)//batch_sz)
```

```python
def visualize_seg(img, gt_mask):
    fig , ax = plt.subplots(1,3,figsize=(10,10), subplot_kw={
                         'xticks': (), 'yticks': ()})

    ax[0].imshow(img)
    ax[0].set_title("Orignal Image")

    pred_mask = model.predict(np.expand_dims(img, 0))
    pred_mask = np.argmax(pred_mask, axis=-1)
    pred_mask = pred_mask[0]
    
    rgb_mask = np.apply_along_axis(map_class_to_rgb, -1, np.expand_dims(pred_mask, -1))

    ax[1].imshow(rgb_mask)
    ax[1].set_title("Predicted Mask")             

    ax[2].imshow((gt_mask).astype(np.uint8))
    ax[2].set_title("Ground truth")
```
